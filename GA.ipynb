# Pre Processing

```python
def PreProcesing(): # filing nans
    dataset['DiabetesMellitus'].fillna('0',inplace=True)
    dataset['ChronicKidneyDisease'].fillna('0',inplace=True)
    dataset['Anemia'].fillna('0',inplace=True)
    dataset['Depression '].fillna('0',inplace=True)
    dataset['ChronicObstructivePulmonaryDisease'].fillna('0',inplace=True)
```

```python
PreProcesing()
```

```python
def encounter(dataset): # dealing with encounterID column
    
    encounterid = []
    for x in dataset["EncounterId"]:
        if type(x) == str:
            x = 0
        if x > 10000000000:
            x = 0    
        encounterid.append(x)
    
    dataset['EncounterId'] = encounterid
```

```python
encounter(dataset)
```

```python
def encoding(dataset): # encoding strings
    labelE = LabelEncoder()

    columns = ['DischargeDisposision','EncounterId','Gender','Race','DiabetesMellitus','ChronicKidneyDisease','Anemia','Depression ','ChronicObstructivePulmonaryDisease','ReadmissionWithin_90Days']
    dataset[columns] = dataset[columns].apply(labelE.fit_transform)
```

# Fitness Function

```python
def FitnessFunction(ffX , ffY): # LGBM checks accuracies
    
    ffx_train , ffx_test , ffy_train , ffy_test = train_test_split(ffX ,ffY)
    model = lgb.LGBMClassifier(learning_rate = 0.09, max_depth=25, random_state = 42)
    model.fit(ffx_train, ffy_train)
    prediction = model.predict(ffx_test)
    accuracy = accuracy_score(ffy_test,prediction)*100

    return accuracy
```

# Population Initialization

```python
patterens = []
```

```python
def population_initialization(): # making intial population
    sX = dataset.drop('ReadmissionWithin_90Days' , axis = 1)
    sY = dataset['ReadmissionWithin_90Days']
    
    columns = dataset.columns
    selected = random.randint(2, size=(56))
    
    modified_dataset = pd.DataFrame()
    for i in range(56):
        if selected[i] == 1:
            modified_dataset[columns[i]] = sX[columns[i]]
    patterens.append(selected) 
    
    #population_initialization_display(selected)
    return FitnessFunction(modified_dataset , sY)
```

# Selection

```python
accuracies = []
```

```python
def Sorting(accuracies , patterens): # Sorting both accuracies , patterens list
    for i in range(len(accuracies)): 
        for j in range(len(accuracies)):
            temp = 0
            if accuracies[i] < accuracies[j]:
                temp = accuracies[i]
                accuracies[i] = accuracies[j]
                accuracies[j] = temp

                temp = 0
                temp = patterens[i]
                patterens[i] = patterens[j]
                patterens[j] = temp 
```

```python
def generation(): # making first generation
    for i in range(60):
        acc = population_initialization()
        accuracies.append(acc)    
```

```python
def poping(pop_elements): #v poping low accuracy members
    for i in range(pop_elements):
        accuracies.pop(0)
        patterens.pop(0)
```

```python
def print_selected():
    for i in range(len(patterens)):
        print("Selected Patteren " , i , " : " , patterens[i])
```

# Crossover

```python
def crossover_accuracy(crossovered): # calculating accuracy of crossovered childrens
    cX = dataset.drop('ReadmissionWithin_90Days' , axis = 1)
    cY = dataset['ReadmissionWithin_90Days']
    
    modified_dataset = pd.DataFrame()
    columns = dataset.columns

    for i in range(56):
        if crossovered[i] == 1:
            modified_dataset[columns[i]] = cX[columns[i]]

    return FitnessFunction(modified_dataset , cY)
```

```python
def crossover(): # crossover implementation
    crossovered = []
    
    for i in range(0 , len(accuracies) - 1 , 2):
        list1 = patterens[i]
        list2 = patterens[i + 1]
        
        rand_num = random.randint(0,56)
        
        list3 = []
        list4 = []
        
        for j in range(rand_num):
            list3.append(list1[j])
            list4.append(list2[j])

        for k in range(rand_num , 56):
            list3.append(list2[k])
            list4.append(list1[k])
            
        patterens.append(list3)
        patterens.append(list4)
        crossovered.append(list3)
        crossovered.append(list4)        
        
        #print_crossovers(rand_num , list1, list2, list3, list4)
    for x in range(len(crossovered)):
         accuracies.append(crossover_accuracy(crossovered[x]))
```

```python
def print_crossovers(rand_num , list1 , list2 , list3, list4):
    print("Splitting Point : " , rand_num)
    print("Parent 1 : " , list1)
    print("Parent 2 : " , list2)
    print("Parent 3 : " , list3)
    print("Parent 4 : " , list4)
    print("\n")
```

# Mutation

```python
def mutation_check(): # checking need for mutation
    rand_num = random.randint(0,100)
    
    if rand_num < 50:
        rand_num = random.randint(0 , 60)
        mutation(rand_num)
```

```python
def mutation(index): # mutation implementation
    rand_num = random.randint(0 , 56)

    #print("Index : " , index)
    #print("Changing At Column Number : " , rand_num)
    
    #print("Non-Mutated patteren : " ,  patterens[index])
    if patterens[index][rand_num] == 1:
        patterens[index][rand_num] = 0
        
    elif patterens[index][rand_num] == 0:
        patterens[index][rand_num] = 1
        
    #print("Mutated patteren : " ,  patterens[index])
```

```python
def same(list1 , list2): # checking if 2 lists are same
    counter = 0
    
    for i in range(len(list1)):
        if list1[i] == list2[i]:
            counter += 1
            
    if counter == len(list1):
        return True
    else:
        return False
```

```python
def mutation_checksame(): # checking need of mutation after mutation
    for i in range(0 , len(patterens)):
        if (i+1) < len(patterens):
            for j in range(i + 1 , len(patterens)):
                if same(patterens[i] , patterens[j]) == True:
                    mutation(j)
                    mutation_checksame()
```

# Genetic Algorithim

```python
def print_best(best_accuracy , best_chromosome):
    for i in range(len(best_accuracy)): # Printing
        print("Accuracy : " , best_accuracy[i])
        print("Patteren : " , best_chromosome[i])
        print("\n")
```

```python
def make_dataframe(best_accuracy , best_chromosome): # making dataframe of patterens and accuracies
    column_names = dataset.columns
    column_names = column_names[0:56]
    
    patterens_dataframe = pd.DataFrame(best_chromosome , columns=column_names)
    patterens_dataframe['Acuuracies'] = best_accuracy
    patterens_dataframe.head(5)
    return patterens_dataframe
```

```python
def GeneticAlgorithim(): # GA
    
    best_chromosome = []
    best_accuracy = []
    
    generation() # Selection
    Sorting(accuracies , patterens)
    poping(30)
    
    for i in range(200): # Termination Criteria 1
        
        crossover() # Crossover
        mutation_check()# mutation
        mutation_checksame# if any combination is same
        
        Sorting(accuracies , patterens)
        poping(30)
        
        best_chromosome.append(patterens[0])
        best_accuracy.append(accuracies[0])
        
        if i > 20: # Termination Criteria 2
            if best_accuracy[i] == best_accuracy[i - 1]:
                break
        
        if accuracies[0] == 90: # Termination Criteria 3
            break
            
    Sorting(best_accuracy , best_chromosome)
    #print_best(best_accuracy , best_chromosome)
    return make_dataframe(best_accuracy , best_chromosome)
```
